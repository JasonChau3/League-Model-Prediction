# final model

```python
#final model.


ohe = Pipeline(steps = [('ohe', OneHotEncoder(handle_unknown='ignore'))])

scaler = Pipeline(steps = [('scaler', StandardScaler())])
preproc = ColumnTransformer(transformers = 
                            [('scaled',scaler, ['blueGoldDiff', 'blueExperienceDiff','blueTotalGold','redTotalGold',
                                               'blueTotalExperience','redTotalExperience','blueGoldPerMin',
                                               'redGoldPerMin', 'blueTotalMinionsKilled','redTotalMinionsKilled'])]
                              ,remainder = 'passthrough')


pl = Pipeline(steps = [('preprocessing', preproc), ('regressor', xgb.XGBClassifier())])

pl.fit(X_train,y_train)
```

```python
vals = (pl.predict(X_test) == y_test).value_counts()
vals[True] / (vals[True] + vals[False])
```

```python
#data cleaning 
#reviewscorevalue is missing alot of value, can see if we want to use this

df['neighbourhood_group_cleansed']
```

```python
#most important features: beds, bathroom, area, review_scores, cancellation policy

#test to see which is better zip code or cities
def clean_na(daf):
    daf['bedrooms'] = daf['bedrooms'].fillna(method='ffill')
    daf['bathrooms'] =daf['bathrooms'].fillna(method='ffill')
    daf['city'] = daf['city'].fillna(method='ffill')
    #so apparantly city is much better than zipcode somehow...
    #daf['zipcode'] = daf['zipcode'].astype(str)
    #daf['zipcode'] =daf['zipcode'].fillna(method='ffill')
    daf['property_type'] =daf['property_type'].fillna(method='ffill')
    daf['room_type'] =daf['room_type'].fillna(method='ffill')
    daf['review_scores_rating'] =daf['review_scores_rating'].fillna(method='bfill')
    daf['host_identity_verified'] =daf['host_identity_verified'].fillna(method='ffill')
    
    #daf['host_listing_count'] = daf['host_listings_count'].fillna(method='ffill')
    daf['neighbourhood_cleansed'] =daf['neighbourhood_cleansed'].fillna(method='ffill')
    #daf['neighbourhood_group_cleansed'] =daf['neighbourhood_group_cleansed'].fillna(method='ffill')
    daf['room_type'] =daf['room_type'].fillna(method='bfill')
    daf['reviews_per_month'] =daf['reviews_per_month'].fillna(method='bfill')
    
    #daf['host_verifications'] =daf['host_verifications'].fillna(method='bfill')
    daf['guests_included'] = daf['guests_included'].fillna(method ='ffill')
    
    daf['accommodates'] = daf['accommodates'].fillna(method ='ffill')

    #daf['review_scores_cleanliness'] =daf['review_scores_cleanliness'].fillna(method='bfill')
    #daf['review_scores_location'] =daf['review_scores_location'].fillna(method='bfill')
    #daf['review_scores_checkin'] =daf['review_scores_checkin'].fillna(method='bfill')
    #daf['review_scores_communication'] =daf['review_scores_communication'].fillna(method='bfill')

    daf['cancellation_policy'] =daf['cancellation_policy'].fillna(method='ffill')
    
    #daf['calculated_host_listings_count'] = daf['calculated_host_listings_count'].fillna(method='ffill')
    daf['number_of_reviews'] =daf['number_of_reviews'].fillna(method='bfill')
    daf['host_neighbourhood'] =daf['host_neighbourhood'].fillna(method='bfill')

    daf['extra_people'] =daf['extra_people'].fillna(method='bfill')
    daf['extra_people'] = daf['extra_people'].str.strip('$')
    
    daf['host_response_rate'] = daf['host_response_rate'].fillna(method ='bfill')
    daf['host_response_rate'] = daf['host_response_rate'].str.strip('%')
    daf['host_response_rate'] = pd.to_numeric(daf['host_response_rate'])
       
    
    daf['extra_people'] = pd.to_numeric(daf['extra_people'])
    daf['room_type'] = daf['room_type'].astype(str)
    daf['host_is_superhost'] = daf['host_is_superhost'].fillna(method='ffill')
    
    return daf[['host_response_rate','accommodates','property_type','reviews_per_month','guests_included','city','host_neighbourhood','number_of_reviews','extra_people','cancellation_policy','bedrooms','bathrooms', 'host_is_superhost','room_type', 'review_scores_rating','host_identity_verified']]
    
```

```python
cleaned_df = clean_na(df)
display(cleaned_df)
cleaned_df.dtypes
```

```python
test_df = pd.read_csv('test.csv')

clean_test = clean_na(test_df)

clean_test
```

```python
ohe = Pipeline(steps = [('ohe', OneHotEncoder(handle_unknown='ignore'))])

scaler = Pipeline(steps = [('scaler', StandardScaler())])
preproc = ColumnTransformer(transformers = 
                            [('zipcode', ohe, ['room_type','cancellation_policy','city',
                                               'host_identity_verified', 'host_is_superhost','property_type',
                                               'room_type','host_neighbourhood']),('scaled',scaler, ['number_of_reviews',
                                                                                'extra_people','guests_included','reviews_per_month', 'host_response_rate',                                                        
                                                                                'bedrooms','bathrooms','review_scores_rating','accommodates'])],remainder = 'passthrough')


pl = Pipeline(steps = [('preprocessing', preproc), ('regressor', xgb.XGBRegressor())])

pl.fit(cleaned_df,df['price'])

```

```python

#xgboost
vals = pl.predict(clean_test)

print(vals)
```

```python
output_df = pd.DataFrame()
output_df['Id'] = test_df['id']
output_df['Predicted'] = vals
output_df.to_csv('simple_linear_regression_baseline.csv', index = False)
```
